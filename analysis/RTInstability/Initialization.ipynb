```python
import numpy
import matplotlib.pyplot as pyplot
```

```python
xlow = -0.5
xhigh = 0.5
ylow = -2
yhigh = 2

x = numpy.linspace(xlow,xhigh,20)
y = numpy.linspace(ylow,yhigh,80)

xmesh, ymesh = numpy.meshgrid(x,y)

amp = 0.05
eta = 0.01
```

```python
dfun = numpy.tanh(amp*numpy.cos(2*numpy.pi*(xmesh+xlow))/(numpy.sqrt(2*eta)) - ymesh)
```

```python
pyplot.figure()
pyplot.contourf(xmesh,ymesh,dfun)
pyplot.axis("equal")
```

```python
pyplot.figure()
pyplot.contour(xmesh,ymesh,dfun,levels=0)
pyplot.axis("equal")
```
