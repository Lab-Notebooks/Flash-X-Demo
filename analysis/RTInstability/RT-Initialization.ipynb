```python
import numpy
import matplotlib.pyplot as pyplot
```

```python
x = numpy.linspace(0,1,20)
y = numpy.linspace(0,4,80)

xmesh, ymesh = numpy.meshgrid(x,y)
```

```python
dfun = -numpy.tanh(ymesh - 2 - 0.1*numpy.cos(2*numpy.pi*(xmesh)))
```

```python
pyplot.figure()
pyplot.contourf(xmesh,ymesh,dfun)
pyplot.axis("equal")
```

```python
pyplot.figure()
pyplot.contour(xmesh,ymesh,dfun,levels=0)
pyplot.axis("equal")
```
